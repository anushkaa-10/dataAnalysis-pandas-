```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
## totalbills and tip
sns.relplot(x='total_bill', y = 'tip', data = df)
```

```python
##hue
sns.relplot(x='total_bill', y = 'tip',hue = 'smoker', data = df)
```

```python
##style
sns.relplot(x='total_bill', y = 'tip',hue='smoker',style='sex', data = df)
```

```python
##categorical
sns.catplot(x='day',y='tip',kind='strip',data=df)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=df)
```

```python
sns.boxplot(df['tip'])
```

```python
sns.catplot(x='day',y='tip',kind='box',data=df)
```

```python
sns.catplot(x='day',y='tip',kind='box',hue='sex',data=df)
```

```python
sns.catplot(x='day',y='tip',kind='violin',hue='sex',data=df)
```

```python
sns.catplot(x='day',y='tip',kind='violin',hue='sex',split=True,data=df)
```

```python
##barplot
sns.catplot(x='smoker',y='total_bill',kind='bar',data=df)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',estimator=np.median,data=df)
```

```python
##countplots
sns.catplot(x='sex',kind='count', data=df)
```

```python
sns.jointplot(x='total_bill', y='tip',data=df)
```

```python
sns.jointplot(x='total_bill', y='tip',kind='hex',data=df)
```

```python
sns.jointplot(x='total_bill', y='tip',kind='kde',data=df)
```

```python
sns.jointplot(x='total_bill', y='tip',kind='reg',data=df)
```

```python
titanic= sns.load_dataset('titanic')
titanic.head()
```

```python
##distplot
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic.isna().sum()
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```
